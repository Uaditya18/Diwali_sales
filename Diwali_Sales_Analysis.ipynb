```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_sales=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
df_sales
```

```python
df_sales.shape
```

```python
df_sales.head()
```

```python
df_sales.info()
```

```python
df_sales.drop(["Status","unnamed1"],axis=1,inplace=True)
df_sales.info()
```

```python
pd.isnull(df_sales).sum()
```

```python
df_sales.shape
```

```python
df_sales.dropna(inplace=True)
```

```python
df_sales.shape
df_sales.info()
```

```python
df_sales['Amount']=df_sales['Amount'].astype('int64')
```

```python
df_sales.info()
```

```python
df_sales.rename(columns={'Marital_Status':'Shadi'})
```

```python
df_sales.describe()
```

Gender


```python
sns.countplot(x='Gender',data=df_sales)
```

```python
ax = sns.countplot(x='Gender',data=df_sales)
ax.set_title("count plot between male and female envoled in the sales")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df_sales.groupby(["Gender"],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
```

```python
sales_gen=df_sales.groupby(["Gender"],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

age

```python
ax =sns.countplot(x='Age Group',data=df_sales,hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df_sales.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values('Age Group',ascending=False)
sns.barplot(x='Age Group',y='Amount',data= sales_age,)
```

```python
sales_state=df_sales.groupby(['State'],as_index=False)['Orders'].sum().sort_values('Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x="State",y= 'Orders',data= sales_state)
```

```python
sales_state=df_sales.groupby(['State'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x="State",y= 'Amount',data= sales_state)
```

```python
ax=sns.countplot(x="Marital_Status",data=df_sales,hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars);
```

```python
sales_state=df_sales.groupby(['Marital_Status'],as_index=False)['Orders'].sum().sort_values('Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(x="Marital_Status",y= 'Orders',data= sales_state)
```

Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax= sns.countplot(x='Occupation',data=df_sales,hue='Gender')

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_state=df_sales.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x="Occupation",y= 'Amount',data= sales_state)
```
